```python
%gui tk

# ==============================
# IMPORT LIBRARIES
# ==============================

import pandas as pd
import nltk
import tkinter as tk
from tkinter import messagebox, ttk
from nltk.corpus import stopwords
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.linear_model import LogisticRegression

nltk.download('punkt')
nltk.download('stopwords')

# ==============================
# OWNER PASSWORD
# ==============================

OWNER_PASSWORD = "admin123"

# ==============================
# TRAINING DATA
# ==============================

positive_reviews = [
"pizza was amazing","food was delicious","very tasty meal",
"burger was juicy","sandwich was delicious","pancakes was soft","biryani tasted great"
]

negative_reviews = [
"food was bad","pizza was cold","burger tasted horrible",
"sandwich was toasted hard","meal was terrible","pancakes was burnt","biryani was oily"
]

reviews = positive_reviews + negative_reviews
labels = ["Positive"]*len(positive_reviews) + ["Negative"]*len(negative_reviews)

df = pd.DataFrame({"Review":reviews,"Sentiment":labels})

# ==============================
# TEXT PROCESSING
# ==============================

stop_words = set(stopwords.words("english"))

def preprocess(text):
    text = text.lower()
    tokens = nltk.word_tokenize(text)
    tokens = [w for w in tokens if w.isalpha() and w not in stop_words]
    return " ".join(tokens)

df["Clean"] = df["Review"].apply(preprocess)

# ==============================
# TRAIN MODEL
# ==============================

vectorizer = TfidfVectorizer()
X = vectorizer.fit_transform(df["Clean"])
y = df["Sentiment"]

model = LogisticRegression()
model.fit(X,y)

# ==============================
# FOOD ITEMS + ICONS
# ==============================

food_items = {
"Pizza":"🍕",
"Burger":"🍔",
"Sandwich":"🥪",
"Pancakes":"🥞",
"Biryani":"🍗"
}

items = list(food_items.keys())

# ==============================
# REVIEW STORAGE
# ==============================

review_data = {}

for item in items:
    review_data[item] = {
        "Positive":0,
        "Negative":0,
        "Ratings":[],
        "Reviews":[]
    }

# ==============================
# GUI START
# ==============================

root = tk.Tk()
root.title("Restaurant Review Sentiment Analysis System")
root.geometry("950x550")
root.configure(bg="#f0f8ff")

title = tk.Label(root,
text="Restaurant Review Sentiment Analysis System",
font=("Arial",16,"bold"),
bg="#f0f8ff",
fg="#2c3e50")

title.pack(pady=10)

main_frame = tk.Frame(root,bg="#f0f8ff")
main_frame.pack()

# ==============================
# CUSTOMER PANEL
# ==============================

customer_frame = tk.Frame(main_frame,
bd=2,
relief="solid",
padx=20,
pady=20,
bg="#e8f6f3")

customer_frame.grid(row=0,column=0,padx=20)

tk.Label(customer_frame,
text="Customer Panel",
font=("Arial",14,"bold"),
bg="#e8f6f3").pack(pady=10)

tk.Label(customer_frame,
text="Select Food Item",
bg="#e8f6f3").pack()

selected_item = tk.StringVar(root)
selected_item.set(items[0])

menu = tk.OptionMenu(customer_frame,selected_item,*items)
menu.pack(pady=5)

# FOOD ICON DISPLAY
food_icon_label = tk.Label(customer_frame,font=("Arial",40),bg="#e8f6f3")
food_icon_label.pack()

def update_food_icon(*args):
    item = selected_item.get()
    food_icon_label.config(text=food_items[item])

selected_item.trace("w",update_food_icon)
update_food_icon()

tk.Label(customer_frame,
text="Write Review",
bg="#e8f6f3").pack(pady=5)

review_entry = tk.Entry(customer_frame,width=30)
review_entry.pack()

# ==============================
# STAR RATING
# ==============================

tk.Label(customer_frame,
text="Select Rating",
bg="#e8f6f3").pack(pady=5)

rating_var = tk.IntVar(value=0)

star_frame = tk.Frame(customer_frame,bg="#e8f6f3")
star_frame.pack()

stars=[]

def set_rating(val):
    rating_var.set(val)
    update_stars()

def update_stars():
    rating = rating_var.get()
    for i in range(5):
        if i < rating:
            stars[i].config(text="★",fg="gold")
        else:
            stars[i].config(text="☆",fg="black")

for i in range(5):
    star = tk.Label(star_frame,text="☆",font=("Arial",25),bg="#e8f6f3")
    star.pack(side="left",padx=2)
    star.bind("<Button-1>",lambda e,i=i:set_rating(i+1))
    stars.append(star)

# ==============================
# AVERAGE RATING LABEL
# ==============================

avg_label = tk.Label(customer_frame,
text="Average Rating: ⭐ 0",
font=("Arial",12,"bold"),
bg="#e8f6f3")

avg_label.pack(pady=10)

# ==============================
# SUBMIT REVIEW
# ==============================

def submit_review():

    review = review_entry.get()
    item = selected_item.get()
    rating = rating_var.get()

    if review=="" or rating==0:
        messagebox.showwarning("Warning","Enter review and rating")
        return

    clean = preprocess(review)
    vector = vectorizer.transform([clean])
    prediction = model.predict(vector)[0]

    review_data[item][prediction]+=1
    review_data[item]["Ratings"].append(rating)
    review_data[item]["Reviews"].append((review,"★"*rating,prediction))

    ratings = review_data[item]["Ratings"]
    avg = round(sum(ratings)/len(ratings),2)

    avg_label.config(text=f"{item} Average Rating ⭐ {avg}")

    messagebox.showinfo("Review Submitted",
    f"{item} Review Added\nRating: {'★'*rating}\nSentiment: {prediction}")

    review_entry.delete(0,tk.END)
    rating_var.set(0)
    update_stars()

tk.Button(customer_frame,
text="Submit Review",
bg="#27ae60",
fg="white",
font=("Arial",10,"bold"),
command=submit_review).pack(pady=10)

# ==============================
# OWNER LOGIN PANEL
# ==============================

login_frame = tk.Frame(main_frame,
bd=2,
relief="solid",
padx=20,
pady=20,
bg="#fdebd0")

login_frame.grid(row=0,column=1,padx=20)

tk.Label(login_frame,
text="Owner Login",
font=("Arial",14,"bold"),
bg="#fdebd0").pack(pady=10)

tk.Label(login_frame,
text="Enter Passkey",
bg="#fdebd0").pack()

pass_entry = tk.Entry(login_frame,show="*")
pass_entry.pack()

# ==============================
# CLEAR DATA
# ==============================

def clear_data(tree):

    confirm = messagebox.askyesno("Confirm","Clear all review data?")

    if confirm:

        for item in items:
            review_data[item]["Positive"]=0
            review_data[item]["Negative"]=0
            review_data[item]["Ratings"].clear()
            review_data[item]["Reviews"].clear()

        for row in tree.get_children():
            tree.delete(row)

        messagebox.showinfo("Cleared","All data deleted")

# ==============================
# OWNER PANEL
# ==============================

def owner_panel():

    owner = tk.Toplevel(root)
    owner.title("Owner Dashboard")
    owner.geometry("750x450")

    tree = ttk.Treeview(owner,
    columns=("Food","Rating","Review","Sentiment"),
    show="headings")

    tree.heading("Food",text="Food Item")
    tree.heading("Rating",text="Star Rating")
    tree.heading("Review",text="Written Review")
    tree.heading("Sentiment",text="Sentiment")

    tree.pack(fill="both",expand=True)

    for item in items:
        for r in review_data[item]["Reviews"]:
            tree.insert("",tk.END,values=(item,r[1],r[0],r[2]))

    # ANALYTICS
    stats_frame = tk.Frame(owner)
    stats_frame.pack(pady=10)

    most_positive = max(items,key=lambda x:review_data[x]["Positive"])
    most_negative = max(items,key=lambda x:review_data[x]["Negative"])

    tk.Label(stats_frame,
    text=f"⭐ Most Loved Food: {most_positive}",
    font=("Arial",11,"bold")).pack()

    tk.Label(stats_frame,
    text=f"⚠ Most Complained Food: {most_negative}",
    font=("Arial",11,"bold")).pack()

    tk.Button(owner,
    text="Clear All Data",
    bg="#e74c3c",
    fg="white",
    command=lambda:clear_data(tree)).pack(pady=10)

# ==============================
# LOGIN CHECK
# ==============================

def open_owner_panel():

    password = pass_entry.get()

    if password == OWNER_PASSWORD:
        owner_panel()
    else:
        messagebox.showerror("Access Denied","Incorrect Password")

tk.Button(login_frame,
text="Login",
bg="#2980b9",
fg="white",
font=("Arial",10,"bold"),
command=open_owner_panel).pack(pady=10)

# ==============================
# RUN GUI
# ==============================

root.mainloop()
```
