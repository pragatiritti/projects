```python
# -----------------------------
# 1. Import Libraries
# -----------------------------
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.preprocessing import LabelEncoder
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score

# -----------------------------
# 2. Create Dataset (Based on Your Image)
# -----------------------------
data = {
    "Employee_ID": [1, 2, 3, 4, 5, 6],
    "Income": ["high", "high", "high", "low", "low", "low"],
    "Credit_Score": ["good", "good", "poor", "good", "poor", "poor"],
    "Employment": ["employed", "unemployed", "employed", "employed", "employed", "unemployed"],
    "Loan_Approval": ["Yes", "Yes", "No", "Yes", "No", "No"]
}

df = pd.DataFrame(data)

print("Dataset:\n")
print(df)

# -----------------------------
# 3. Convert Text to Numbers
# -----------------------------
le = LabelEncoder()

for column in ["Income", "Credit_Score", "Employment", "Loan_Approval"]:
    df[column] = le.fit_transform(df[column])

# -----------------------------
# 4. Split Data
# -----------------------------
X = df[["Income", "Credit_Score", "Employment"]]
y = df["Loan_Approval"]

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=1)

# -----------------------------
# 5. Train Model
# -----------------------------
model = DecisionTreeClassifier()
model.fit(X_train, y_train)

# -----------------------------
# 6. Accuracy
# -----------------------------
y_pred = model.predict(X_test)
print("\nModel Accuracy:", accuracy_score(y_test, y_pred))

# -----------------------------
# 7. Dashboard (Visualization)
# -----------------------------
plt.figure()
df["Loan_Approval"].value_counts().plot(kind="bar")
plt.title("Loan Approval Count")
plt.xlabel("Loan Approval (0 = No, 1 = Yes)")
plt.ylabel("Count")
plt.show()

plt.figure()
df["Income"].value_counts().plot(kind="bar")
plt.title("Income Distribution")
plt.xlabel("Income (0 = high, 1 = low)")
plt.ylabel("Count")
plt.show()

plt.figure()
df["Credit_Score"].value_counts().plot(kind="bar")
plt.title("Credit Score Distribution")
plt.xlabel("Credit Score (0 = good, 1 = poor)")
plt.ylabel("Count")
plt.show()

plt.figure()
df["Employment"].value_counts().plot(kind="bar")
plt.title("Employment Status Distribution")
plt.xlabel("Employment (0 = employed, 1 = unemployed)")
plt.ylabel("Count")
plt.show()
```
