```python
import random
import time

# -----------------------------
# 1. Sensor Simulation Classes
# -----------------------------

class LiDAR:
    def get_distance(self):
        # Simulate distance to nearest obstacle (meters)
        return random.uniform(1, 100)

class Radar:
    def get_speed_of_object(self):
        # Simulate speed of nearby object (km/h)
        return random.uniform(0, 120)

class Camera:
    def detect_objects(self):
        objects = ["None", "Pedestrian", "Vehicle", "Traffic Light - Red", "Traffic Light - Green"]
        return random.choice(objects)

class GPS:
    def get_location(self):
        return (random.uniform(12.0, 13.0), random.uniform(77.0, 78.0))


# -----------------------------
# 2. Sensor Fusion Module
# -----------------------------

class SensorFusion:
    def __init__(self, lidar, radar, camera, gps):
        self.lidar = lidar
        self.radar = radar
        self.camera = camera
        self.gps = gps

    def get_fused_data(self):
        data = {
            "distance": self.lidar.get_distance(),
            "object_speed": self.radar.get_speed_of_object(),
            "object_type": self.camera.detect_objects(),
            "location": self.gps.get_location()
        }
        return data


# -----------------------------
# 3. Decision-Making System
# -----------------------------

class DecisionSystem:
    def decide(self, data):
        distance = data["distance"]
        obj_type = data["object_type"]
        speed = data["object_speed"]

        if obj_type == "Traffic Light - Red":
            return "STOP"

        elif obj_type == "Pedestrian" and distance < 20:
            return "STOP"

        elif distance < 10:
            return "BRAKE"

        elif speed > 80:
            return "SLOW DOWN"

        else:
            return "MOVE FORWARD"


# -----------------------------
# 4. Autonomous Vehicle System
# -----------------------------

class AutonomousVehicle:
    def __init__(self):
        self.lidar = LiDAR()
        self.radar = Radar()
        self.camera = Camera()
        self.gps = GPS()

        self.fusion = SensorFusion(self.lidar, self.radar, self.camera, self.gps)
        self.decision_system = DecisionSystem()

    def run(self):
        print("🚗 Autonomous Vehicle Started...\n")

        for _ in range(10):  # Run for 10 cycles (simulation)
            data = self.fusion.get_fused_data()

            decision = self.decision_system.decide(data)

            print("📡 Sensor Data:", data)
            print("🧠 Decision:", decision)
            print("-" * 50)

            time.sleep(1)


# -----------------------------
# 5. Run Simulation
# -----------------------------

if __name__ == "__main__":
    vehicle = AutonomousVehicle()
    vehicle.run()
```
