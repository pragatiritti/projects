```python
import re
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.linear_model import LogisticRegression

# ---------------------------
# Rule Based Logic System
# ---------------------------

def check_degree(text):
    degrees = ["btech", "b.e", "bsc computer science", "cs degree", "computer science"]
    return any(degree in text.lower() for degree in degrees)

def check_experience(text):
    match = re.search(r'(\d+)\s+years', text.lower())
    if match:
        years = int(match.group(1))
        return years >= 2
    return False

def check_python(text):
    return "python" in text.lower()

def rule_engine(text):
    degree = check_degree(text)
    experience = check_experience(text)
    python_skill = check_python(text)

    state = {
        "degree": degree,
        "experience": experience,
        "python": python_skill
    }

    if degree and experience and python_skill:
        decision = "Shortlisted ✅"
    else:
        decision = "Rejected ❌"

    return state, decision


# ---------------------------
# Simple ML Model
# ---------------------------

train_data = [
    "BTech computer science 3 years python developer",
    "BSc computer science 5 years python",
    "Mechanical engineer 1 year matlab",
    "Civil engineer 2 years autocad"
]

labels = [1, 1, 0, 0]

vectorizer = CountVectorizer()
X = vectorizer.fit_transform(train_data)

model = LogisticRegression()
model.fit(X, labels)

def ml_predict(text):
    X_test = vectorizer.transform([text])
    prediction = model.predict(X_test)[0]

    if prediction == 1:
        return "AI Prediction: Suitable"
    else:
        return "AI Prediction: Not Suitable"
```

```python
resume = "BTech Computer Science with 3 years experience in Python and cybersecurity"

state, decision = rule_engine(resume)
ai_result = ml_predict(resume)

print("Decision:", decision)
print("AI Result:", ai_result)
print("Logic State:", state)
```
