```python
import random
import time

# -----------------------------
# 1. Product Database
# -----------------------------

products = {
    "Laptop": {"price": 50000, "stock": 10},
    "Phone": {"price": 20000, "stock": 20},
    "Headphones": {"price": 2000, "stock": 30},
    "Shoes": {"price": 3000, "stock": 15},
    "Watch": {"price": 2500, "stock": 25}
}

# Simulated user history
user_history = ["Laptop", "Phone"]


# -----------------------------
# 2. Dynamic Pricing System
# -----------------------------

class DynamicPricing:
    def update_price(self, product_name):
        product = products[product_name]

        demand = random.randint(1, 10)  # Simulated demand
        stock = product["stock"]

        # Pricing logic
        if demand > 7:
            product["price"] *= 1.10   # Increase price
        elif stock < 5:
            product["price"] *= 1.15   # Low stock → higher price
        else:
            product["price"] *= 0.95   # Discount

        product["price"] = round(product["price"], 2)
        return product["price"]


# -----------------------------
# 3. Recommendation System
# -----------------------------

class RecommendationSystem:
    def recommend(self, history):
        recommendations = []

        for item in history:
            if item == "Laptop":
                recommendations.append("Headphones")
            elif item == "Phone":
                recommendations.append("Watch")
            elif item == "Shoes":
                recommendations.append("Watch")

        return list(set(recommendations))


# -----------------------------
# 4. Inventory Management
# -----------------------------

class InventoryManager:
    def update_stock(self, product_name):
        product = products[product_name]

        # Simulate sales
        sold = random.randint(0, 3)
        product["stock"] -= sold

        # Replenishment logic
        if product["stock"] < 5:
            product["stock"] += 10
            print(f"🔄 Restocked {product_name}")

        return product["stock"]


# -----------------------------
# 5. Online Retail System
# -----------------------------

class OnlineRetailSystem:
    def __init__(self):
        self.pricing = DynamicPricing()
        self.recommendation = RecommendationSystem()
        self.inventory = InventoryManager()

    def run(self):
        print("🛒 Online Retail System Running...\n")

        for _ in range(5):  # simulate 5 cycles
            print("📦 Product Updates:\n")

            for product in products:
                new_price = self.pricing.update_price(product)
                new_stock = self.inventory.update_stock(product)

                print(f"{product} → Price: ₹{new_price}, Stock: {new_stock}")

            # Recommendations
            recs = self.recommendation.recommend(user_history)
            print("\n🎯 Recommended Products:", recs)

            print("\n" + "=" * 50 + "\n")
            time.sleep(1)


# -----------------------------
# 6. Run System
# -----------------------------

if __name__ == "__main__":
    system = OnlineRetailSystem()
    system.run()
```
