```python
import re

# -------------------------
# Step 1: Extract Facts
# -------------------------

def extract_facts(resume_text):
    facts = set()
    text = resume_text.lower()

    if "computer science" in text or "btech cs" in text:
        facts.add("cs_degree")

    match = re.search(r'(\d+)\s+years', text)
    if match:
        years = int(match.group(1))
        if years >= 2:
            facts.add("experience_2plus")

    if "python" in text:
        facts.add("python_skill")

    return facts


# -------------------------
# Step 2: Forward Chaining Engine
# -------------------------

def forward_chaining(facts):

    rules = [
        ({"cs_degree", "experience_2plus", "python_skill"}, "eligible"),
    ]

    inferred = set()
    changed = True

    while changed:
        changed = False
        for conditions, conclusion in rules:
            if conditions.issubset(facts) and conclusion not in inferred:
                inferred.add(conclusion)
                changed = True

    return inferred


# -------------------------
# Step 3: Final Decision
# -------------------------

def resume_scanner(resume_text):

    facts = extract_facts(resume_text)
    inferred = forward_chaining(facts)

    if "eligible" in inferred:
        decision = "Shortlisted ✅"
    else:
        decision = "Rejected ❌"

    return facts, inferred, decision
```

```python
resume = "Diploma in Mechanical Engineering with 1 year experience in AutoCAD"

facts, inferred, decision = resume_scanner(resume)

print("Facts:", facts)
print("Inferred:", inferred)
print("Decision:", decision)
```

```python
resume = "BTech Computer Science with 3 years experience in Python"

facts, inferred, decision = resume_scanner(resume)

print("Facts:", facts)
print("Inferred:", inferred)
print("Decision:", decision)
```
